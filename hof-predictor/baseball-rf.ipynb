```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Data//baseball-data-final.csv')
```

```python
df.head()
```

```python
df2 = df
```

```python
df2["OBP"] = (df2["hits"]+df2["walks"]+df2["hbp"])/(df2["ab"]+df2["walks"]+df2["hbp"]+df2["sacf"])
```

```python
df2.head()
```

```python
df2["SLG"] = ((df2["singles"]+(df2["doubles"]*2)+df2["triples"]*3)+(df2["hr"]*4))/df2["ab"]
```

```python
df2["AVG"] = (df2["hits"]/df2["ab"])
```

```python
df2.head()
```

```python
df2.loc[df2["playerID"]=="brettge01"]
```

```python
pos = df2.loc[df2["mfp"]!= "PITCHER"]
```

```python
pos.head()
```

```python
pos.info()
```

```python
pos.isna().any()
```

```python
pos.fillna(0, inplace=True)
```

```python
pos.isna().any()
```

```python
final = pos.drop(columns = ['playerID','nameFirst','nameLast','ab','singles','doubles','triples','hbp','sacf'])
```

```python
final.head()
```

```python
final.info()
```

```python
bats = pd.get_dummies(final['bats'], drop_first=True)
throws = pd.get_dummies(final['throws'], drop_first=True)
mfp = pd.get_dummies(final['mfp'], drop_first=True)
```

```python
final.drop(columns=['bats','throws','mfp'], inplace=True)
```

```python
final = pd.concat([final,bats,throws,mfp], axis = 1)
```

```python
final.head()
```

```python
final.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(final.drop('inducted',axis=1), 
                                                    final['inducted'], test_size=0.30, 
                                                    random_state=101)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators = 100)
rfc.fit(X_train, y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
print(confusion_matrix(y_test, rfc_pred))
```

```python
print(classification_report(y_test, rfc_pred))
```

```python
merged = X_test
```

```python
merged['inducted']=(y_test)
merged['prediction']=(rfc_pred)
```

```python
merged.head()
```

```python
merged['playerID']=(df['playerID'])
merged['nameFirst']=(df['nameFirst'])
merged['nameLast']=(df['nameLast'])
```

```python
merged.head()
```

```python
merged.to_csv("Data//player_predictions_rf.csv")
```
